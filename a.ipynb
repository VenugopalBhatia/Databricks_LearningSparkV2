```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("BasicsOfPySpark").getOrCreate()
```

```python
rdd = spark.sparkContext.parallelize([1,2,3,4,5])
rdd.count()
```

```python
data = [('James','','Smith','1991-04-01','M',3000),
  ('Michael','Rose','','2000-05-19','M',4000),
  ('Robert','','Williams','1978-09-05','M',4000),
  ('Maria','Anne','Jones','1967-12-01','F',4000),
  ('Jen','Mary','Brown','1980-02-17','F',-1)
]
```

```python
schema = ["firstname","middlename","lastname","dob","gender","salary"]
df = spark.createDataFrame(data,schema = schema)
```

```python
df.show()
```

```python
df.groupBy('gender').avg('salary').show()
```

```python
df.sort(df['salary'].desc()).show()
```

```python
from pyspark.sql.functions import col
```

```python
df.sort(col('salary').desc()).show()
```

```python
df.withColumn('rich',expr('salary>3000')).where('rich').show()
```

```python
df.withColumn('year',split('dob','-')[0])\
    .withColumn('month',split('dob','-')[1])\
    .withColumn('date',split('dob','-')[2]).show()
```

```python
df.createOrReplaceTempView('df_view')
```

```python
spark.sql("Select * from df_view order by dob desc").show()
```

```python
df.select('*')\
.where(year(col('dob'))>'1970').show()
```

```python
countLen = udf(lambda x:len(x))
```

```python
df.withColumn('dtcount', countLen('dob'))\
.withColumn('firstNameCount',countLen('firstname'))\
.sort(col('firstNameCount').desc())\
.show()
```
